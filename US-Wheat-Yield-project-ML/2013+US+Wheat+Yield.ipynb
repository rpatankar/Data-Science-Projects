```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# created a function for training a regression model and calculating the mean squared error and variance. 
#This function takes in a list of strings.

def train_and_test(cols):
    # Split into features & target.
    features = wheat_filtered[cols]
    target = wheat_filtered["Yield"]
    # Fit model.
    lr = LinearRegression()
    lr.fit(features, target)
    # Make predictions on training set.
    predictions = lr.predict(features)
    # Compute MSE and Variance.
    mse = mean_squared_error(wheat_filtered["Yield"], predictions)
    variance = np.var(predictions)
    return (mse, variance)

# example: humidity_mse, humidity_var = train_and_test(["humidity"])
```

```python
# Now will use this function to to build several regression models examining the effects of the above factors on yield.
# Next step will be to use cross-validation to test the models since no separate test data exists.


one_mse_var = train_and_test(["apparentTemperatureMax"])
two_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin"])
three_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover"])
four_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint"])
five_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity"])
six_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity'])
seven_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax'])
eight_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax',  "precipProbability"])
nine_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation"])
ten_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation",  "pressure"])
eleven_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity",'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation",  "pressure",  "temperatureMax"])
twelve_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity",'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation",  "pressure",  "temperatureMax", "temperatureMin"])
thirteen_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation",  "pressure",  "temperatureMax", "temperatureMin", "windSpeed"])
fourteen_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation",  "pressure",  "temperatureMax", "temperatureMin", "windSpeed", "NDVI"])
#fifteen_mse_var = train_and_test(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation", "pressure",  "temperatureMax", "temperatureMin", "windSpeed", "NDVI", "windSpeed"])
# windspeed is repeated ..... 
```

```python
# Since there is no separate TEST dataset, I will attempt a cross-validation method to compare in sample versus out of sample errors

from sklearn.cross_validation import KFold
from sklearn.metrics import mean_squared_error
import numpy as np

def train_and_cross_val(cols):
    features = wheat_filtered[cols]
    target = wheat_filtered["Yield"]
    
    variance_values = []
    mse_values = []
    
    # KFold instance.
    kf = KFold(n=len(wheat_filtered), n_folds=10, shuffle=True, random_state=3)
    
    # Iterate through over each fold.
    for train_index, test_index in kf:
        # Training and test sets.
        X_train, X_test = features.iloc[train_index], features.iloc[test_index]
        y_train, y_test = target.iloc[train_index], target.iloc[test_index]
        
        # Fit the model and make predictions.
        lr = LinearRegression()
        lr.fit(X_train, y_train)
        predictions = lr.predict(X_test)
        
        # Calculate mse and variance values for this fold.
        mse = mean_squared_error(y_test, predictions)
        var = np.var(predictions)

        # Append to arrays to do calculate overall average mse and variance values.
        variance_values.append(var)
        mse_values.append(mse)
   
    # Compute average mse and variance values.
    avg_mse = np.mean(mse_values)
    avg_var = np.mean(variance_values)
    return(avg_mse, avg_var)
        
one_mse, one_var = train_and_cross_val(["apparentTemperatureMax"])
two_mse, two_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin"])
three_mse, three_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover"])
four_mse, four_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint"])
five_mse, five_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity"])
six_mse, six_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity'])
seven_mse, seven_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax'])
eight_mse, eight_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax',  "precipProbability"])
nine_mse, nine_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation"])
ten_mse, ten_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation",  "pressure"])
eleven_mse, eleven_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity",'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation",  "pressure",  "temperatureMax"])
twelve_mse, twelve_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity",'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation",  "pressure",  "temperatureMax", "temperatureMin"])
thirteen_mse, thirteen_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation",  "pressure",  "temperatureMax", "temperatureMin", "windSpeed"])
fourteen_mse, fourteen_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation",  "pressure",  "temperatureMax", "temperatureMin", "windSpeed", "NDVI"])
#fifteen_mse, fifteen_var = train_and_cross_val(["apparentTemperatureMax", "apparentTemperatureMin", "cloudCover", "dewPoint", "humidity", 'precipIntensity', 'precipIntensityMax', "precipProbability", "precipAccumulation", "pressure",  "temperatureMax", "temperatureMin", "windSpeed", "NDVI", "windSpeed"])
```
